## Simple Linear Regression

## Step 1. Import the necessary libraries

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

## Step 2. Import the dataset

```python
df = pd.read_csv("50_Startups.csv")
```

```python
df
```

## Step 3 . Check the State Column

```python
df['State'].value_counts()
```

## Step 4 . Splitting dataset in to input and output

```python
x = df.drop(['Profit'], axis=1)
```

```python
y = df['Profit']
y
```

## Step 5 . Convert state Column into Numeric Column

## Step 5.1 .  Perform Transformation

```python
x1 = pd.get_dummies(x,columns=['State'],drop_first=True)
x1
```

## Step 6 . Dummy variable trap

```python
# Already Performed using | drop_first =True
```

## Step 7 Splitting dataset in to Train and Test

```python
x_train, x_test, y_train, y_test = train_test_split(x1,y,test_size=0.2, random_state=42)
print("Training Shape: " , x_train.shape)
print("Testing Shape:  " , x_test.shape)
```

## Step 8  Import LinearRegression model from linear_model family 

```python
from sklearn.linear_model import LinearRegression
```

## Step 9  Fit the data

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

## Step 10  Predict the data

```python
y_pre = model.predict(x_test)
y_pre
```

## Step 11  Display Result

```python
# y_test and y_predict
```

```python
sns.scatterplot(x=x_test['Administration'], y=y_test)
sns.scatterplot(x=x_test['Administration'], y=y_pre)
```

## RSS

```python
np.sum((y_test.values - y_pre ) ** 2)
```

```python
len(y_test)
```

```python
from sklearn.metrics import mean_squared_error , r2_score
```

```python
mean_squared_error(y_test.values , y_pre) * len(y_pre)
```

```python
len(y_test)
```

## R Square

```python
r2_score(y_test,y_pre)
```

## Now use Polynomial Regression on Position_Salaries dataset

```python
df = pd.read_csv("Position_Salaries.csv")
df
```

```python
sns.scatterplot(x=df['Level'],y =df['Salary'])
```

```python
x = df.iloc[::,1:2:]
x
```

```python
y = df['Salary']
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
print("Training Shape: " , x_train.shape)
print("Testing Shape:  " , x_test.shape)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2)
```

```python
x1 = poly.fit_transform(x_train)
x1
```

```python
model1 = LinearRegression()
model1.fit(x1,y_train)
```

```python
y_poly_pre = model1.predict(poly.fit_transform(x_test))
y_poly_pre
```
